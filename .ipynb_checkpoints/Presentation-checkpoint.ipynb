```python
import pandas as pd
from decimal import Decimal, getcontext
import matplotlib.pyplot as plt

# Set decimal precision
getcontext().prec = 20
n=94
i=7
# Step 1: Read CSV as strings to avoid float conversion
df = pd.read_csv('%s%i_%istars_ex_red_mu.csv'%(data_out+process_step[5],i,n), dtype = str)

# Step 2: Convert all applicable values to Decimal
def safe_decimal(x):
    try:
        return Decimal(x)
    except:
        return x  # non-numeric or header values remain as-is

df = df.applymap(safe_decimal)

# Step 3: Do your high-precision calculations (e.g., variance)
def decimal_variance(series):
    numeric_series = [x for x in series if isinstance(x, Decimal)]
    n = Decimal(len(numeric_series))
    mean = sum(numeric_series) / n
    return sum((x - mean) ** 2 for x in numeric_series) / (n - 1)

# Example: Add a row for variance
df.loc['var'] = [decimal_variance(df[col]) if all(isinstance(x, Decimal) for x in df[col]) else None for col in df.columns]

# Step 4: Convert Decimal to float ONLY for plotting
df_float = df.applymap(lambda x: float(x) if isinstance(x, Decimal) else x)

# Step 5: Plot the data (e.g., all rows except 'var')
df_float.drop('var', errors='ignore').T.plot(kind='bar')
plt.title("Precise Data Plot")
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```
