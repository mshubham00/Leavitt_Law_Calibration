```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation

# Step 1: Generate some data
x = np.linspace(0, 2*np.pi, 50)
y = 2 * np.sin(x)

fig, ax = plt.subplots()

# Step 3: Initialize your lines
line, = ax.plot(x, y)

def update(num):
    # Step 4: Update function to update your lines with new data points
    line.set_ydata((2 * np.sin(x + num / 50)))  
    return line,

# To just draw static plot call plt.show() instead of following line
ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 100), blit=True)

plt.show()
```
